- `noise_factor_monitoreo`: 0.03
- `noise_factor_prenada`: 0.05

- `clip_prenada_range`: [20.0, 50.0] L

````python
print("Aplicando Data Augmentation para balancear el dataset...")

def augment_animal_sequence(df_animal, n_augmentations=5):
    """
    Augmenta datos de un animal espec铆fico con reglas realistas por estado
    """
    augmented_data = []
    estado = df_animal['Estado de Producci贸n'].iloc[0]

    for _ in range(n_augmentations):
        df_aug = df_animal.copy()

        if estado == 'En Monitoreo':
            noise_factor = 0.03
            df_aug['Producci贸n de leche ayer'] *= (1 + np.random.normal(0, noise_factor, len(df_aug)))
            df_aug['Producci贸n de leche ayer'] = df_aug['Producci贸n de leche ayer'].clip(upper=25.0)
            df_aug['Producci贸n media diaria 煤ltimos 7 d铆as'] *= (1 + np.random.normal(0, noise_factor, len(df_aug)))
            df_aug['Producci贸n media diaria 煤ltimos 7 d铆as'] = df_aug['Producci贸n media diaria 煤ltimos 7 d铆as'].clip(upper=25.0)

        elif estado == 'Pre帽ada':
            noise_factor = 0.05
            df_aug['Producci贸n de leche ayer'] *= (1 + np.random.normal(0, noise_factor, len(df_aug)))
            df_aug['Producci贸n de leche ayer'] = df_aug['Producci贸n de leche ayer'].clip(lower=20.0, upper=50.0)
            df_aug['Producci贸n media diaria 煤ltimos 7 d铆as'] *= (1 + np.random.normal(0, noise_factor, len(df_aug)))
            df_aug['Producci贸n media diaria 煤ltimos 7 d铆as'] = df_aug['Producci贸n media diaria 煤ltimos 7 d铆as'].clip(lower=20.0, upper=50.0)

        elif estado == 'Seca':
            df_aug['Producci贸n de leche ayer'] = 0.0
            df_aug['Producci贸n media diaria 煤ltimos 7 d铆as'] = 0.0

        else:
            noise_factor = 0.05
            df_aug['Producci贸n de leche ayer'] *= (1 + np.random.normal(0, noise_factor, len(df_aug)))
            df_aug['Producci贸n media diaria 煤ltimos 7 d铆as'] *= (1 + np.random.normal(0, noise_factor, len(df_aug)))

        shift_days = np.random.randint(-3, 4)
        df_aug['D铆as en orde帽o'] = (df_aug['D铆as en orde帽o'] + shift_days).clip(lower=0)

        if len(df_aug) > 3:
            mask = np.random.choice([True, False], size=len(df_aug), p=[0.2, 0.8])
            for col in ['Producci贸n de leche ayer', 'Producci贸n media diaria 煤ltimos 7 d铆as']:
                if df_aug[col].sum() > 0:
                    df_aug.loc[mask, col] = df_aug[col].interpolate(method='linear')

        augmented_data.append(df_aug)

    return pd.concat(augmented_data, ignore_index=True)

def balance_dataset(df, target_col='Estado de Producci贸n', min_samples=10000):
    """
    Balancea el dataset con augmentaci贸n realista
    """
    class_counts = df[target_col].value_counts()
    balanced_dfs = []

    print("=" * 80)
    print("BALANCEO DE DATASET CON REGLAS REALISTAS")
    print("=" * 80)
    print(f"\nDistribuci贸n ORIGINAL:")
    for estado, count in class_counts.items():
        print(f"  {estado}: {count} registros")

    for estado in df[target_col].unique():
        df_estado = df[df[target_col] == estado]
        current_count = len(df_estado)

        if current_count < min_samples:
            n_augmentations = (min_samples // current_count) + 1
            print(f"\nProcesando '{estado}': {current_count}  {current_count * n_augmentations} registros")

            animals = df_estado['N煤mero del animal'].unique()
            for animal in animals:
                df_animal = df_estado[df_estado['N煤mero del animal'] == animal]
                augmented = augment_animal_sequence(df_animal, n_augmentations)
                balanced_dfs.append(augmented)
        else:
            print(f"\n '{estado}' ya tiene suficientes muestras: {current_count}")
            balanced_dfs.append(df_estado)

    result_df = pd.concat(balanced_dfs, ignore_index=True)

    print("\n" + "=" * 80)
    print("DISTRIBUCIN FINAL:")
    print("=" * 80)
    final_counts = result_df[target_col].value_counts()
    for estado, count in final_counts.items():
        print(f"  {estado}: {count} registros")

    return result_df

# Aplicar data augmentation al dataset original
df_balanced = balance_dataset(df)

print("\n" + "=" * 80)
print("VALIDACIN DE RANGOS DE PRODUCCIN POR ESTADO")
print("=" * 80)

for estado in df_balanced['Estado de Producci贸n'].unique():
    df_estado = df_balanced[df_balanced['Estado de Producci贸n'] == estado]
    prod_ayer = df_estado['Producci贸n de leche ayer']
    prod_media = df_estado['Producci贸n media diaria 煤ltimos 7 d铆as']

    print(f"\n{estado}:")
    print(f"  Producci贸n ayer - Min: {prod_ayer.min():.2f}, Max: {prod_ayer.max():.2f}, Media: {prod_ayer.mean():.2f}")
    print(f"  Producci贸n media 7d - Min: {prod_media.min():.2f}, Max: {prod_media.max():.2f}, Media: {prod_media.mean():.2f}")

print("\n" + "=" * 80)
print("VERIFICACIN DE CASOS PROBLEMTICOS")
print("=" * 80)

en_monitoreo_alta = df_balanced[
    (df_balanced['Estado de Producci贸n'] == 'En Monitoreo') &
    (df_balanced['Producci贸n de leche ayer'] > 25)
]
print(f"Registros 'En Monitoreo' con producci贸n > 25 litros: {len(en_monitoreo_alta)}")

prenada_muy_baja = df_balanced[
    (df_balanced['Estado de Producci贸n'] == 'Pre帽ada') &
    (df_balanced['Producci贸n de leche ayer'] < 15)
]
print(f"Registros 'Pre帽ada' con producci贸n < 15 litros: {len(prenada_muy_baja)}")

print(f"\nDataset balanceado listo para MLP")
print(f"Total de registros: {len(df_balanced)}")
````
